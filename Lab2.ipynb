```python
data = pd.read_csv('Sales.csv')
data
```

```python
month = data['month_number']
profit = data['total_profit']
plt.figure(figsize=(10, 6))
plt.plot(month, profit, marker='o', linestyle='-', color='b')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Over Months')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(month,profit,linestyle='dotted',color='r',marker='*',markerfacecolor='b',linewidth=3,label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.title('Total Profit Over Months')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
